```python
import pandas as pd
```

```python
Demo= pd.read_csv('Demografia2021.csv')
Demo
```

```python
Demo.sort_values(by='Poblacion', ascending=False)
```

```python
# Paso 1: Ordenar los pesos (Poblaci√≥n) de menor a mayor
df_sorted = Demo.sort_values(by='Poblacion', ascending=False)
df_sorted

```

```python
# Paso 2: Asegurarnos de que 'Poblacion' sea de tipo entero
df_sorted['Poblacion'] = df_sorted['Poblacion'].astype(int)

# Paso 3: Calcular las diferencias entre poblaciones consecutivas
df_sorted['diff'] = df_sorted['Poblacion'].diff().abs()  # Diferencias absolutas entre las poblaciones consecutivas
```

```python
covid = pd.read_csv('Covid2021.csv')
covid
```

```python
# Crear el DataFrame

# Paso 1: Calcular la Poblaci√≥n Susceptible a partir de los Nuevos Casos y la Tasa de Incidencia
covid['Susceptible'] = (covid['New_Cases'] * 1000) / covid['Incident_Rate']

# Mostrar los resultados
covid
```

```python
# Guardar el DataFrame en un archivo CSV
covid.to_csv('Covid2021_1.csv', index=False)
```

```python

covFip= covid.copy()

# Lista de FIPS que deseas filtrar
fips_deseados = [41, 40, 9, 49, 19, 32, 5, 28, 20]

# Filtrar el DataFrame para obtener solo los FIPS deseados
covFip = covFip[covFip['FIPS'].isin(fips_deseados)]

# Mostrar el nuevo DataFrame
covFip
```
